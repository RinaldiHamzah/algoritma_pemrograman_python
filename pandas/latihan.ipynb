```python
import pandas as pd
df = pd.read_excel("nilai_mhs.xlsx")
df
```

```python
def nilai_huruf(nilai):
    if nilai>=80:
        return 'A'
    elif nilai>=70:
        return 'B'
    elif nilai>=60:
        return 'C'
    elif nilai>=40:
        return 'D'
    else:
        return 'E'
df['skor'] = df['Nilai'].apply(nilai_huruf)
print(df)
```

```python
# filter
print("----------filter----------")
x=df[df['Nim'] == 1001]
print(x)
```

```python
y = df.groupby('Nim')['Nilai'].mean()
print(y)
```
