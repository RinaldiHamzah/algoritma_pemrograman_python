Menggunkan Library Pandas

```python
import pandas as pd
```

```python
data = {
    'tanggal': [
        '2024-01-01', '2024-01-01',
        '2024-01-02', '2024-01-02',
        '2024-02-01', '2024-02-01',
        '2024-03-01', '2024-03-01'
    ],
    'produk': [
        'Buku', 'Pulpen',
        'Buku', 'Penghapus',
        'Pulpen', 'Buku',
        'Buku', 'Pulpen'
    ],
    'kategori': [
        'ATK', 'ATK', 'ATK', 'ATK',
        'ATK', 'ATK', 'ATK', 'ATK'
    ],
    'harga': [
        10000, 3000, 10000, 2000,
        3000, 10000, 10000, 3000
    ],
    'jumlah': [
        2, 5, 1, 3, 4, 2, 5, 6
    ]
}

df = pd.DataFrame(data)
df['tanggal'] = pd.to_datetime(df['tanggal'])
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df["total_penjualan"] = df["harga"] * df["jumlah"]
df["bulan"] = df["tanggal"].dt.month
```

```python
df["total_penjualan"].sum()
```

```python
df.groupby("produk") ["total_penjualan"].sum()
```

```python
df.groupby("produk") ["jumlah"].sum().sort_values(ascending=False)
```

```python
df.groupby("bulan") ["total_penjualan"].sum()
```

```python
df.groupby("produk") ["total_penjualan"].sum().sort_values(ascending=False)
```

```python
ranking = (
    df.groupby("produk") ["total_penjualan"]
    .sum()
    .rank(ascending=False, method="dense")
)
ranking
```

```python
laporan = (
    df.groupby(["bulan", "produk"])
    .agg(
        total_jumlah= ("jumlah", "sum"),
        total_penjumlahan=("total_penjualan", "sum")
    )
    .reset_index()
    )
```
