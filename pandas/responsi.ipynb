```python
#filtering data
def filter_data(usia):
    if usia >=18:
        return "kawin"
    else:
        return "belum lawin"
df['status'] = df['usia'].apply(filter_data)  
```

```python
#jumlah kecamatan
jumlah_kecamatan = df["kecamatan"].value_counts()
print(jumlah_kecamatan)
```

```python
#kelompok usia
def kelompok(usia):
    if usia<=17:
        return 'anak'
    elif usia<=30:
        return 'dewasa muda'
    elif usia<=50:
        return 'dewasa'
    else:
        return 'lansia'
df['umur'] = df['usia'].apply(kelompok)
print(df)
```
